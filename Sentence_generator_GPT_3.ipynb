```python
from transformers import pipeline
```

```python
generator = pipeline('text-generation',model='EleutherAI/gpt-neo-2.7B')
```

```python
def generated_text(prompt):
  response = generator(prompt,max_length=150,do_sample=True,temperature=0.9)
  if isinstance(response, list) and response:
      generated_text = response[0].get('generated_text', '')
      return generated_text

  return response
```

```python
generated_text('AI vs Human')
```

```python
def textbox(name):
    return generated_text(name)

demo = gr.Interface(fn=textbox, inputs="textbox", outputs="textbox",title='Sentence Generator using GPT-3')

if __name__ == "__main__":
    demo.launch()
```
